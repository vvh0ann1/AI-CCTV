```python
import os
import glob
from IPython.display import Image, display

# 디렉토리 찾기
exp_directories = glob.glob('/content/drive/MyDrive/yolov5/runs/detect/exp*')

# 디렉토리가 존재하지 않으면 예외 발생
if not exp_directories:
    raise Exception('디렉토리를 찾을 수 없습니다.')

# 가장 최근의 디렉토리를 선택
latest_exp_directory = max(exp_directories, key=os.path.getctime)

# 선택된 디렉토리 내의 이미지를 출력
for imageName in glob.glob(os.path.join(latest_exp_directory, '*.jpg')):
    display(Image(filename=imageName))
    print("\n")
```
